### **training ml(s) pt. 1**

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
```

```python
x_train = seatrain[['Latitude', 'Longitude', 'NDSI']]
y_train = seatrain['label']

x_test = seatest[['Latitude', 'Longitude', 'NDSI']]
y_test = seatest['label']

model_v1 = DecisionTreeClassifier()
model_v1.fit(x_train, y_train) ## ^


model_v1_viz = export_graphviz(model_v1,
                         out_file = None, feature_names = ['lat', 'long', 'ndsi'],
                         class_names = ['water', 'plastic', 'driftwood', 'seaweed',
                                        'pumice', 'sea snot', 'sea snot'],
                         filled = True)

graph = graphviz.Source(model_v1_viz)
graph.render("graphviz_decision_tree", format = "png")
graph
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = model_v1.predict(x_test) ##^

cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot = True, cmap = 'Blues', xticklabels = ['water', 'plastic', 'driftwood', 'seaweed', 'pumice', 'sea snot'], yticklabels=['water', 'plastic', 'driftwood', 'seaweed', 'pumice', 'sea snot'])
plt.title('Model 1')
plt.xlabel('Pred Label')
plt.ylabel('True Label')
plt.show()
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model_v1.predict(x_test) ## making predictions from

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the Decision Tree model: {accuracy:.2f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier ##^^
```

```python
features = seadata_all[['NDSI', 'MNDWI', 'NDWI', 'OSI', 'FDI', 'B1']]
target_y = seadata_all['label']


x_train, x_test, y_train, y_test = train_test_split(features, target_y, test_size = 0.3, random_state = 13)


model_vml = DecisionTreeClassifier(random_state = 13)
model_vml.fit(x_train, y_train) ## ^


model_vml_viz = export_graphviz(model_vml,
                         out_file = None, feature_names = ['ndsi', 'mndwi', 'ndwi',
                                                          'osi', 'fdi', 'band_1'],
                         class_names = ['water', 'plastic', 'driftwood', 'seaweed',
                                        'pumice', 'sea snot', 'sea foam'],
                         filled = True)

graph = graphviz.Source(model_vml_viz)
graph.render("graphviz_decision_tree", format = "png")
graph
```

```python
y_pred = model_vml.predict(x_test) ## from modelv6

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Model: {accuracy:.2f}")
```

```python
y_pred = model_vml.predict(x_test) ##^

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize = (8, 6))
sns.heatmap(cm, annot = True, cmap = 'Blues', xticklabels = ['water', 'plastic', 'driftwood', 'seaweed', 'pumice', 'sea snot', 'sea foam'],
                                              yticklabels=['water', 'plastic', 'driftwood', 'seaweed', 'pumice', 'sea snot', 'sea foam'])
plt.title('Decision Tree, Model Performance', pad = 20)
plt.xlabel('Predicted Class')
plt.ylabel('Actual Label')
plt.show()
```

```python
y_pred = model_vml.predict(x_test) ##^

cm = confusion_matrix(y_test, y_pred, normalize = 'true')

plt.figure(figsize = (8, 6))
sns.heatmap(cm, annot = True, cmap = 'Blues', xticklabels = ['Water', 'Plastic', 'Driftwood', 'Seaweed', 'Pumice', 'Sea Snot', 'Sea Foam'],
                                              yticklabels=['Water', 'Plastic', 'Driftwood', 'Seaweed', 'Pumice', 'Sea Snot', 'Sea Foam'],
                                              fmt='.2f',
                                              annot_kws={"size": 9})
plt.title('Decision Tree Model Performance', pad = 20)
plt.xlabel('Predicted Class')
plt.ylabel('Actual Label')

plt.tick_params(axis='x', which='both', bottom=False, top=False)
plt.tick_params(axis='y', which='both', left=False, right=False)

plt.xticks(fontsize = 9)
plt.yticks(fontsize = 9)

plt.show()
```
